```python
def stopwordslist(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
    return stopwords
stopwords = stopwordslist('F:/Jupyter/--NLP/顺丰项目/stopwords.txt')
```

```python
from collections import Counter
def getTfidfFromDir(dir_path):
    t1 = time.time()
    file_list = os.listdir(dir_path)
    doc_num = len(file_list) #文档数
    file_all_words = []  # 所有文档中所有的的词
    file_unique_words = []  # 每个文档的词的set合并得到
    for file in file_list:
        filepath = dir_path+'/'+file
        with open(filepath,'r',encoding='utf-8') as f:
            words = jieba.lcut(f.read())
            words = [x for x in words if x not in stopwords]  ## 去除乱七八糟的停用词
            unique_words = list(set(words))
            file_all_words += words
            file_unique_words += unique_words
        print(filepath,'finished!')
    t2 = time.time()
    print("分词耗时：",t2-t1,'s')
    print("文档唯一词量：",len(list(set(file_all_words))))
    freq_dic = Counter(file_all_words)        ## 每个词：在所有文档中出现的次数
    in_how_many_dic = Counter(file_unique_words) ## 每个词：在多少个文档中出现
    t3 = time.time()
    print("统计词频耗时：",t3-t2,'s')
    return in_how_many_dic,freq_dic,doc_num
```

```python
in_how_many_dic,freq_dic,doc_num = getTfidfFromDir('C:/Users/x1c/Documents/WeChat Files/LIghTGBY/Files/sf_cvs/sf_cvs')
```

```python
import math
def calculateWeight(freq_in_doc,words_num_in_doc,total_doc_num,in_how_many_doc,freq_in_wiki):
    """
    parameters：
    freq_in_doc -- 该词在某个文档中的次数
    words_num_in_doc -- 该文档的总词数
    total_doc_num -- 总文档数
    in_how_many_doc -- 某词出现在多少的文档中
    freq_in_wiki -- 在wiki中的流行度
    
    """
    tf = freq_in_doc/words_num_in_doc
    idf = math.log10(total_doc_num/in_how_many_doc)
    wiki_weight = 1/(math.log10(freq_in_wiki+1))
    
    final_weight = tf*idf*wiki_weight
    print("tf:",tf)
    print("idf:",idf)
    print("wiki_weight:",wiki_weight)
    return final_weight
```

```python
def getImportantWordsFromDoc(text,total_doc_num):
    words = list(set(jieba.lcut(text)))
    words = [x for x in words if x not in stopwords]
    all_words = jieba.lcut(text)
    words_weights_dic = {}
    for word in words:
        freq_in_doc = all_words.count(word)
        words_num_in_doc = len(all_words)
        total_doc_num = total_doc_num
        in_how_many_doc = in_how_many_dic[word]
        freq_in_wiki = wiki_word_freq_dic[word] if word in list(wiki_word_freq_dic.keys()) else 1
        print('-------',word)
        weight = calculateWeight(freq_in_doc,words_num_in_doc,total_doc_num,in_how_many_doc,freq_in_wiki)
        words_weights_dic[word] = weight
        
    print('***************************************')
    result = sorted(words_weights_dic.items(),key = lambda x:x[1])
    print(result)
#     return words_weights_dic
```

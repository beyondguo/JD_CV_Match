```python
import pandas as pd
from pandas import DataFrame
import numpy
data = DataFrame(pd.read_csv("C:/Users/x1c/Desktop/猎上顺丰简历项目/10月/测试.csv",\
                  engine='python',encoding='utf-8'))
data
```

```python
import re
def beyondSplit(x):
    sentences = re.split("。|，|？|！|：|；|\.|,|\?|\!|:|;|\n|\t",x)
    return sentences
def clean(x):
    x = re.sub('([0-9][、，。\.])','',x).strip()
    x = re.sub('  ','',x)  # 两个空格去掉，但是如果只有一个空格则要保留，因为很多词确实中间需要一个空格
    return x
s = clean("1、擅哈Java Spring\
       a.阿斯蒂芬\
       C，需要")
s
```

```python
jr = data['job_req']
jd = data['job_desc']
print(jr.shape,jd.shape)
jr_sentences = []
jd_sentences = []
for each1,each2 in zip(jr,jd):
    sentences1 = [x for x in beyondSplit(clean(str(each1))) if len(x)>20]
    sentences2 = [x for x in beyondSplit(clean(str(each2))) if len(x)>20]
    jr_sentences = jr_sentences + sentences1
    jd_sentences = jd_sentences + sentences2
print(len(jr_sentences))
print(len(jd_sentences))
jr_sentences = list(set(jr_sentences))
jd_sentences = list(set(jd_sentences))
print(len(jr_sentences))
print(len(jd_sentences))
```

```python
import os
LTP_DATA_DIR = 'F:/MyDownloads/ltp_data_v3.4.0/ltp_data_v3.4.0'  # ltp模型目录的路径
cws_model_path = os.path.join(LTP_DATA_DIR, 'cws.model')  # 分词模型路径
from pyltp import Segmentor # 分词器
from pyltp import Postagger # 词性标注器
from pyltp import Parser # 句法分析器

segmentor = Segmentor()  # 初始化实例
segmentor.load(cws_model_path)  # 加载模型
pos_model_path = os.path.join(LTP_DATA_DIR, 'pos.model')  # 词性标注模型路径，模型名称为`pos.model`
postagger = Postagger() # 初始化实例
postagger.load(pos_model_path)  # 加载模型
par_model_path = os.path.join(LTP_DATA_DIR, 'parser.model')  # 依存句法分析模型路径，模型名称为`parser.model`
parser = Parser() # 初始化实例
parser.load(par_model_path)  # 加载模型

def get_words_and_tags(sentence):
    words = list(segmentor.segment(sentence))
    postags = list(postagger.postag(words))
    return [words,postags]

relation_dic = {'SBV':'主谓关系','VOB':'动宾关系','IOB':'间宾关系','FOB':'前置宾语',\
                'DBL':'兼语','ATT':'定中关系','ADV':'状中结构','CMP':'动补结构',\
                'COO':'并列关系','POB':'介宾关系','LAD':'左附加关系','RAD':'右附加关系','IS':'独立结构','HED':'核心关系','WP':'标点'}

def getParserResult(sentence):
    words,postags = get_words_and_tags(sentence)
    arcs = parser.parse(words, postags)  # 句法分析
    title = "句子：\n"+sentence+'\n VOB:\n'
    print(title)
    results = ''
    for index,arc in enumerate(arcs):
        relation_word = words[arc.head-1] if arc.head>0 else 'ROOT'
        relation = arc.relation
        combination = words[index]+relation_word if index<=arc.head-1 else relation_word+words[index]
        tag = postags[index]
        relation_tag = postags[arc.head-1] if arc.head>0 else 'ROOT'
        
        if relation == 'VOB':
            result = words[index]+'('+tag+')'+'|'+relation_word+'('+relation_tag+')'+'|'+combination
            results = results + result + '\n'
            print(result)
    print('\n-------------')
    return title+results+"\n------------\n"
```

```python
with open('C:/Users/x1c/Desktop/猎上顺丰简历项目/10月/test_vob.txt',encoding='utf-8') as f:
    ss = f.readlines()
    print(len(ss))
    with open('C:/Users/x1c/Desktop/猎上顺丰简历项目/10月/test_vob_results.txt','w') as nf:
        for s in ss:
            results = getParserResult(s)
            nf.write(results)
    #         print(results)
        
parser.release()
segmentor.release()
postagger.release()
```
